```python
import pandas as pd
from random import shuffle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# convert 'C' and 'D' to payoffs
# utilityncomment needed line
insensitive = 0
utility, loss, C, P = 6, 3, 0, 0
# utility=6, loss=1, C=0, P=-2
# utility=6, loss=3, C=6, P=0.

df_payoff = df
for i in range(len(teachers)):
    for j in range(len(teachers)):
        if df_payoff.iloc[i][j] == '0, 0':
            df.iloc[i][j] = 0
        if df_payoff.iloc[i][j] == 'C, C':
            df.iloc[i][j] = utility + C + insensitive  - loss
        if df_payoff.iloc[i][j] == 'C, D':
            df.iloc[i][j] = insensitive  - loss
        if df_payoff.iloc[i][j] == 'D, C':
            df.iloc[i][j] = utility
        if df_payoff.iloc[i][j] == 'D, D':
            df.iloc[i][j] = P
```

```python
# plot results

data = df_payoff.groupby(df_payoff.index).sum()
fig1, ax1 = plt.subplots()
ax1.set_title('Strategies Comparison (insentives=0)') # change the number of incentives to display on the title
ax1.boxplot(data, labels=('altruist', 'ego_centered', 'responsive', 'debatable', 'disappoint'))
plt.show()
```
